```python
#!/usr/bin/env python

# make sure to install these packages before running:
# pip install pandas
# pip install sodapy

import pandas as pd
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.cityofchicago.org", None)

# Example authenticated client (needed for non-public datasets):
# client = Socrata(data.cityofchicago.org,
#                  MyAppToken,
#                  username="user@example.com",
#                  password="AFakePassword")

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("85ca-t3if",     query='SELECT crash_record_id WHERE crash_date > "2023-11-01T17:25:19" :: floating_timestamp ORDER BY crash_date DESC NULL FIRST, crash_record_id ASC NULL LAST')
                     
                     
                     #query='SELECT crash_record_id, crash_date, posted_speed_limit, traffic_control_device, device_condition, weather_condition, lighting_condition, first_crash_type, trafficway_type, alignment, roadway_surface_cond, road_defect, crash_type, date_police_notified, prim_contributory_cause, sec_contributory_cause, street_direction, street_name, beat_of_occurrence, num_units, injuries_total, injuries_fatal, crash_hour, crash_day_of_week, crash_month, :@computed_region_rpca_8um6 ORDER BY crash_date DESC NULL FIRST, crash_record_id ASC NULL LAST')


# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```
